# 5 Additional dataframe update funcitonality 

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19)})
```

```python
df.columns
```

## 5.1 Adding columns

```python
df["c"] = np.arange(20,29) 
df
```

```python
df["d"]=df.a + df.b
df
```

## 5.2 Adding row

```python
df.shape[0] # get number of rows 
```

```python
df.loc[df.shape[0]] = [1,2,3,5]
df
```

## 5.3 Deleting a column

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19), "c": np.arange(20,29)})

df.drop('c', axis=1, inplace=True)
df
```

Some of the padnas dataframe methods accepts the parameter __inplace__ which if sets to True affects the target dataframe. If its set to false a copz is returned

## 5.4 Deleting a row

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19), "c": np.arange(20,29)})

df.drop(2, axis=0, inplace=True)
df
```

## 5.5 Deleting based on condition 

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19), "c": np.arange(20,29)})
df.drop(df[df.a < 3].index) # because 
```

## 5.6 Renaming columns 

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19), "c": np.arange(20,29)})
df.rename(columns={'a': 'New1', 'b': 'Newb'}, inplace=True) # returns copy dont forget the inplace
df
```
