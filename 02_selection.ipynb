## 2.2 Selecting rows
### 2.2.1 Conditions

```python
df.SalePrice > 181500 # returns a bool series
```

```python
df[df.SalePrice > 181500]
```

```python
df[(df.SalePrice < 210000) & (df.SalePrice > 181500)] # Chaning conditions in bool series. Beware the round brackets
# (df.SalePrice < 210000) and (df.SalePrice > 181500) This is error or when using and or not 
```

```python
df[(df.SalePrice < 210000) | (df.SalePrice > 181500)]
```

```python
df[~((df.SalePrice < 210000) & (df.SalePrice > 181500))] # The negation
```

```python
df[df.SalePrice.isin([181500, 210000])] # You can name the values
```

### 2.2.2 Specific delections in conditions 

```python
df.loc[(df.SalePrice < 210000) | (df.SalePrice > 181500), "MSSubClass"]
```

```python
df.iloc[2:4,5:10] # Selection based on the index
```

```python
df.iloc[2:4,:] # reffering to all values 
```
