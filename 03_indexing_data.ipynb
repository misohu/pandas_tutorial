## 3.1 View vs copy of the dataframe
Sometime selection returns a view or a copy of the dataframe.
- __View__: is a selection refering to the original dataframe. Its a reference a pointer
- __Copy__: is a selection refering to the new dataframe.
Beware that the selection returns a reference to df (called view).

Pandas Warns us when we try to updte the copy of the datframe like this.

```python
df=pd.read_csv("house_prices_dataset/train.csv")
temp_df = df[0:1] # Selecting first row
temp_df.Id = 1000
temp_df 
# df
```

This warning is telling you that you are adjusting original dataframe the wrong way. You are chaining the indexers. This can result in nondeterministic behavior (covered bellow). There are two solutions for this:
- If you want to adjust the selection and create a new df from it you need to return the copy, which you can force with the copy function.
- If you want to adjust the original df you can use the loc indexer.

```python
df=pd.read_csv("house_prices_dataset/train.csv")
temp_df = df[0:1].copy()
temp_df.Id = 2000
temp_df
# df
```

```python
df=pd.read_csv("house_prices_dataset/train.csv")
df.loc[0, "Id"] = 3000
df
```

### 3.1.1 When things stop working
Because the `df[df.MSSubClass == 20]` returns a copy of the dataframe and then we adjust the copy !!!.

```python
df=pd.read_csv("house_prices_dataset/train.csv")
df[df.MSSubClass == 20]['MSSubClass'] = 0
df
```

```python
df.loc[df.MSSubClass == 20, 'MSSubClass'] = 0
df
```

Or changing the order !!! Mindblowing

```python
df=pd.read_csv("house_prices_dataset/train.csv")
df['MSSubClass'][df.MSSubClass == 20] = 0
df
```

The conditional select returns the copy not the column select.</br>
BUT WHEN DOES THE PYTHON RETURNS THE VIEW? </br>
Its complex just be sure to use loc or force the copy 

### 3.1.2 Using the copy method
Using this method pandas allows us to create:
- a view if deep = False. 
- a copy if deep = True

```python
df=pd.read_csv("house_prices_dataset/train.csv")
df_view = df.copy(deep=False)
df_copy = df.copy(deep=True) # this is default behavior
```
