# 9.3 Group by columns
Default is axis param set to 0 which means row, but we canset it to columns ... To make thing interesting we can provide agroup by function 

```python
df.head()
```

```python
def starts_with_m(column_name):
    if str.startswith(column_name, "M"):
        return "M_column"
    return "Other_column"
```

```python
grouped = df.groupby(starts_with_m, axis=1)
grouped.first()
```

## 9.4 Group by custom function
Lets say we want to know sum of the lot area and lot frontage for all the MSZonings. To prevent you from this

```python
grp = df.groupby('MSZoning')
```

```python
for key, group in grp:
    print((group.LotArea + group.LotFrontage).sum())
```

Pandas allows you to pass custom function to the grop by like: 

```python
df.groupby('MSZoning').apply(lambda x: (x.LotArea + x.LotFrontage).sum())
```

```python
# or funciton specified ouside like
def area_frontage_sum(group):
    return (group.LotArea + group.LotFrontage).sum()

df.groupby('MSZoning').apply(area_frontage_sum)
```

You can also pass arguments

```python
def area_frontage_sum(group, multiplier):
    return (group.LotArea + group.LotFrontage).sum() * multiplier

df.groupby('MSZoning').apply(area_frontage_sum, multiplier=2)
```
